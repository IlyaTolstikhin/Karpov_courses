```python
import pandas as pd
file = 'gff.tsv'
import seaborn as sns
```

```python
data = []
```

```python
# разделяем строку по табуляции, выбираем нужный столбец, разделяем по точке с запятой, выбираем последний элемент
# аналогично далее
for i in open(file).readlines():
    data.append(i.split('\t')[8].split(';')[-1].split('\n')[0].split('Parent=')[-1].strip(','))
```

```python
data
```

```python
data = pd.DataFrame(data)
data
```

```python
data = data.rename(columns = {'0': 'att'})
```

```python
sns.countplot(data[0])
```
