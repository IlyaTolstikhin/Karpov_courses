```python
path_to_file = "3_taxi_peru.csv"  
taxi = pd.read_csv(path_to_file, sep = ';', parse_dates = ['start_at','end_at','arrived_at'])
```

```python
taxi = taxi.query("(start_type == 'asap') or (start_type == 'reserved') ")
taxi.head()
```

```python
taxi['wait_time'] = (taxi.arrived_at - taxi.start_at).astype('timedelta64[m]')
taxi.head()
```

```python
taxi.query("(start_type == 'reserved') and (wait_time > 0.0) ") \
    .groupby(['driver_id'], as_index = False) \
    .agg({'end_lat' : 'count'}) \
    .sort_values('end_lat', ascending = False) \
    .rename(columns = {'end_lat' : 'count_late'})
```

```python
taxi.start_at.dt.month
```

```python
taxi['month'] = taxi.start_at.dt.month
taxi['weekday'] = taxi.start_at.dt.strftime('%A')
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(taxi.month)
```

```python
taxi.weekday[110:130]
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(taxi.weekday, order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
# mectics
MAU = taxi.groupby(['month'], as_index = False) \
    .agg({'user_id' : pd.Series.nunique})
MAU
```

```python
plt.figure(figsize=(12, 8))
MAU.user_id.plot.line()
```

```python
papka = '/users/Ilya-/data'
type_file = ".csv"
path_to_file = [i for i in os.listdir(papka) if i[-4:] == '.csv']
print(path_to_file)
```

```python
path_to_file = []
for i in os.listdir(papka):
    for j in os.listdir(papka + '/'+ i):
        for k in os.listdir(papka + '/'+ i + '/' + j):
            path_to_file.append(papka + '/'+ i + '/' + j + '/' + k)
```

```python
df = pd.read_csv(path_to_file[0], sep = ',')
df['name'] = path_to_file[0].split('/')[-2]
df['date'] = path_to_file[0].split('/')[-3]
df
```

```python
for i in path_to_file[1:]:
    df1 = pd.read_csv(i, sep = ',')
    df1['name'] = i.split('/')[-2]
    df1['date'] = i.split('/')[-3]
    df = df.append(df1)
df = df.reset_index()
df = df.drop(columns = ['Unnamed: 0', 'index'])
```

```python
df.quantity.sum()
```

```python
df.head()
```

```python
df.groupby(['name'], as_index=False) \
    .agg({'quantity' : 'sum'}) \
    .sort_values('quantity', ascending = False)
```

```python
top_product = df.groupby(['product_id'], as_index=False) \
    .agg({'quantity' : 'sum'}) \
    .sort_values('quantity', ascending = False)
```

```python
sns.barplot(x=top_product.product_id[:9], y=top_product.quantity[:9], order = top_product.product_id[:9].sort_values(ascending = False))
```

```python
weekday_qua = df.groupby(['date'], as_index=False) \
    .agg({'quantity' : 'sum'})
```

```python
# удаляем дубликаты в записях продаж по имени, id и дате, группируем по имени и продукту, считатем количество продаж и отбираем частоту покупок больше 1
name_loyal = df.drop_duplicates(subset=['name', 'product_id', 'date']) \
    .groupby(['name', 'product_id'], as_index=False) \
    .agg({'quantity' : 'count'}) \
    .query('quantity > 1')
# Drop only if duplicates are in 'Date' or 'Last' columns or both of them
name_loyal.drop_duplicates(subset=['name'])
```
