```python
import pandas as pd
```

```python
path_to_file = "2_bookings.csv" 
bookings = pd.read_csv(path_to_file, sep = ';')
bookings_head = bookings[:7]
```

```python
bookings.columns = bookings.columns.str.replace(' ', '_').str.lower()
bookings
```

```python
bookings.query('is_canceled == 0')
```

```python
bookings.nunique()
```

```python
bookings.query('is_canceled == 0') \
    .groupby(['country']) \
    .agg({'hotel' : 'count'}) \
    .sort_values('hotel', ascending = False)
```

```python
b1 = bookings.groupby(['hotel']) \
    .agg({'stays_total_nights' : 'mean'}) \
    .sort_values('hotel', ascending = False)
round(b1.stays_total_nights,2)
```

```python
bookings.query('reserved_room_type != assigned_room_type')
#.agg({'hotel' : 'count'})
#.sort_values('hotel', ascending = False)
```

```python
bookings.query('is_canceled == 1') \
    .groupby(['arrival_date_year', 'arrival_date_month']) \
    .agg({'hotel' : 'count'}) \
    .sort_values('hotel', ascending = False)
```

```python
bookings.agg({'adults' : 'mean', 'children' : 'mean','babies' : 'mean'})
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
b2 = bookings.groupby(['hotel']) \
    .agg({'total_kids' : 'mean'}) \
    .sort_values('total_kids', ascending = False)
round(b2.total_kids,2)
```
